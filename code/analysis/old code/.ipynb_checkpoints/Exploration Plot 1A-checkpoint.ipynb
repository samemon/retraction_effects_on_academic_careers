```python
# Code for creating the bar plot

def create_attrition_barplot(axesi, ticksize, labelsize, attrited_color,
                        nonattrited_color, other_color):
    
    # Only extracting relevant dataframe
    
    # Computing the groups
    df_grouped_values_treatment = df_relevant_w_lastactivity[['MAGAID','YearsActiveNew']]\
                .drop_duplicates()\
                .groupby('YearsActiveNew')['MAGAID'].nunique().reset_index()
    
    # Computing number of authors in our set
    num_authors = df_relevant_w_lastactivity[~df_relevant_w_lastactivity.YearsActiveNew.isna()]\
                            .MAGAID.nunique()
    
    num_authors = df_authors['MAGAID'].nunique()
    
    # Getting Percentage of retracted scientists
    df_grouped_values_treatment['Percentage of Retracted Scientists'] = \
        round(df_grouped_values_treatment['MAGAID']*100.0/num_authors,10)
    
    print("Percentage outside of range",df_grouped_values_treatment[df_grouped_values_treatment['YearsActiveNew'].lt(-5)
                                     | df_grouped_values_treatment['YearsActiveNew'].gt(20)]
         ['Percentage of Retracted Scientists'].sum())
    
    #print(df_grouped_values_treatment['Percentage of Retracted Scientists'].sum())
    
    # Filtering to only include authors within certain years range
    df_grouped_values_treatment = df_grouped_values_treatment[\
                                df_grouped_values_treatment['Percentage of Retracted Scientists'].gt(0) & 
                                df_grouped_values_treatment['YearsActiveNew'].ge(-5) & 
                                df_grouped_values_treatment['YearsActiveNew'].le(20)]
    
    
    
    df_grouped_values_treatment['YearsActiveNew'] = df_grouped_values_treatment['YearsActiveNew'].astype(int)

    # Setting colors for bars
    
    colors = [other_color if (x < -1) else (nonattrited_color if ((x > 0) and (x < 6)) else attrited_color) \
                  for x in df_grouped_values_treatment['YearsActiveNew'].tolist()]


    barplot_treatment = sns.barplot(data=df_grouped_values_treatment, ax=axesi,
                            x='YearsActiveNew', y="Percentage of Retracted Scientists", palette=colors,
                                   saturation=1)

    barplot_treatment.tick_params(labelsize=ticksize)

    for item in barplot_treatment.get_xticklabels():
        item.set_rotation(60)

    for patch in barplot_treatment.patches:
        
        value = patch.get_height()
        if(patch.get_x() in [-0.4, 0.6, 1.6, 2.6, 3.6]):
            barplot_treatment.annotate(format(value, '.1f'), 
                           (patch.get_x() + patch.get_width() / 2., patch.get_height()+1), 
                           ha = 'center', va = 'center', fontsize=labelsize, weight='regular',
                           xytext = (0, 9), color='red',
                           textcoords = 'offset points')
            
        elif value > .1:
            barplot_treatment.annotate(format(value, '.1f'), 
                           (patch.get_x() + patch.get_width() / 2., patch.get_height()+1), 
                           ha = 'center', va = 'center', fontsize=labelsize, weight='regular',
                           xytext = (0, 9), color='blue',
                           textcoords = 'offset points')

    barplot_treatment.set_xlabel('first activity since retraction (years)', 
                          fontsize=labelsize)

    barplot_treatment.set_ylabel('% of retracted authors', fontsize=labelsize)
    
    #barplot_treatment.axvline(5, ymax=0.80, color='black', lw=2, linestyle='--')
    
    barplot_treatment.set_ylim(0,70)
    
    sns.despine()
```

```python
# Code for creating the bar plot

def create_attrition_barplot(axesi, ticksize, labelsize, attrited_color,
                        nonattrited_color, other_color):
    
    # Only extracting relevant dataframe
    
    # Computing the groups
    df_grouped_values_treatment = df_relevant_w_lastactivity[['MAGAID','YearsActiveNew']]\
                .drop_duplicates()\
                .groupby('YearsActiveNew')['MAGAID'].nunique().reset_index()
    
    # Computing number of authors in our set
    num_authors = df_relevant_w_lastactivity[~df_relevant_w_lastactivity.YearsActiveNew.isna()]\
                            .MAGAID.nunique()
    
    num_authors = df_authors['MAGAID'].nunique()
    
    # Getting Percentage of retracted scientists
    df_grouped_values_treatment['Percentage of Retracted Scientists'] = \
        round(df_grouped_values_treatment['MAGAID']*100.0/num_authors,10)
    
    print("Percentage outside of range",df_grouped_values_treatment[df_grouped_values_treatment['YearsActiveNew'].lt(-5)
                                     | df_grouped_values_treatment['YearsActiveNew'].gt(20)]
         ['Percentage of Retracted Scientists'].sum())
    
    #print(df_grouped_values_treatment['Percentage of Retracted Scientists'].sum())
    
    # Filtering to only include authors within certain years range
    df_grouped_values_treatment = df_grouped_values_treatment[\
                                df_grouped_values_treatment['Percentage of Retracted Scientists'].gt(0) & 
                                df_grouped_values_treatment['YearsActiveNew'].ge(-20) & 
                                df_grouped_values_treatment['YearsActiveNew'].le(20)]
    
    
    
    df_grouped_values_treatment['YearsActiveNew'] = df_grouped_values_treatment['YearsActiveNew'].astype(int)

    # Setting colors for bars
    
    colors = [other_color if (x < -1) else (nonattrited_color if (x > 0) else attrited_color) \
                  for x in df_grouped_values_treatment['YearsActiveNew'].tolist()]


    barplot_treatment = sns.barplot(data=df_grouped_values_treatment, ax=axesi,
                            x='YearsActiveNew', y="Percentage of Retracted Scientists", palette=colors,
                                   saturation=1, hatch='/')

    barplot_treatment.tick_params(labelsize=ticksize)

    for item in barplot_treatment.get_xticklabels():
        item.set_rotation(60)

    for patch in barplot_treatment.patches:
        
        value = patch.get_height()
        print(patch.get_x())
        if(patch.get_x() in [16.6, 17.6]):
            barplot_treatment.annotate(format(value, '.1f'), 
                           (patch.get_x() + patch.get_width() / 2., patch.get_height()+1), 
                           ha = 'center', va = 'center', fontsize=labelsize, weight='regular',
                           xytext = (0, 9), color='blue',
                           textcoords = 'offset points')
        elif value > .1:
            barplot_treatment.annotate(format(value, '.1f'), 
                           (patch.get_x() + patch.get_width() / 2., patch.get_height()+1), 
                           ha = 'center', va = 'center', fontsize=labelsize, weight='regular',
                           xytext = (0, 9), color='gray',
                           textcoords = 'offset points')
            

    barplot_treatment.set_xlabel('last activity since retraction (years)', 
                          fontsize=labelsize)

    barplot_treatment.set_ylabel('% of retracted authors', fontsize=labelsize)
    
    barplot_treatment.axvline(18, ymax=0.80, color='black', lw=2, linestyle='--')
    
    barplot_treatment.set_ylim(0,18)
    
    sns.despine()
```

```python
# Data
categories = ['Authors with \nactivity\n post retraction', 'Authors with \nno activity\n post retraction']
values = [(df_authors[df_authors['ActivityOrNo']=='yes activity']['MAGAID'].nunique())*100/df_authors.shape[0], 
          (df_authors[df_authors['ActivityOrNo']=='no activity']['MAGAID'].nunique())*100/df_authors.shape[0]]

plt.rcParams["figure.figsize"] = (2.7,3.7)

# Plotting
fig, ax = plt.subplots()

# Solid yellow bar
solid_bar = ax.bar(categories[0], values[0], color='#F8F8F8', edgecolor='black', width=0.6, label='Solid Yellow')

# Striped yellow bar
striped_bar = ax.bar(categories[1], values[1], color='#F8F8F8', edgecolor='black', hatch='/', width=0.6, label='Yellow with Stripes')

# Function to add annotations
def add_annotations(bars):
    for bar in bars:
        height = round(bar.get_height(),1)
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding annotations
add_annotations(solid_bar)
add_annotations(striped_bar)


# Add labels and legend
ax.set_ylabel('% of retracted authors')
#ax.set_title('Barplot with Two Bars')

sns.despine()

plt.show()
```
